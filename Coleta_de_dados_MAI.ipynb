## Coleta de Dados das fichas do Museu de Arqueologia de Itaipu

O que foi feito?

- Em um primeiro momento com as fichas em mãos houve uma tentativa de extrair os dados diretamente dos pdf´s, que parecia ser o mais simples, no entanto, se mostrou algo muito mais complexo do que o esperado. A partir desse momento foi decidido utilizar os aruqivos .doc uma vez que se mostraram mais maleáveis. Em resumo = word + python + csv <3

Como foi feito?

- Como as fichas não seguem um padrão de metadados, existe uma variabilidade consoante a peça do acervo, foi necessário criar algo controlado que pudesse ser manuseado e alterado de acordo com as necessidades (não se assustem com o tamanho do código)
- Em termos computacionais provavelmente existem diversas formas diferentes de se fazer o que é feito aqui, no entanto essa foi a melhor forma encontrada, para a necessidade que tinhamos.
